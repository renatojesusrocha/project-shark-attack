### Criando função para mudar a coluna YEAR para Int:

```python
def try_int(value):
    try:
        return int(value)
    except:
        return 0  #Função criada para mudar o tipo da coluna. 
```

```python
df.Year = df.Year.map(try_int) #Aplicando função dentro de uma VR. 
```

### Tratando as linhas da coluna Activity 

```python
def tratar_string(row):
    try:
        row = row.lower()
        row = row.strip()
        return row
    except:
        return 'nao_foi'
```
